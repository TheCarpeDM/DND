```python
#  Import dependencies and read the csv.
import pandas as pd 
monsters_df = pd.read_csv("dnd_monsters.csv")
monsters_df.head()
```

```python
#Drop columns we don't need (align & source)
monsters_df = monsters_df.drop(columns=['align', 'source'])
monsters_df.head()
```

```python
#Drop Nan rows ONLY from str
monsters_df = monsters_df.dropna(subset=['str'])
monsters_df.head()
```

```python
#Check data types and changed any if needed
monsters_df.dtypes
```

```python
#CR column should be float but we need to conver fractions
#take from: https://stackoverflow.com/questions/69112191/convert-fractional-values-to-decimal-in-pandas
def convert(frac):
    if '/' in frac:
        num, den = frac.split('/')
        return (int(num)/int(den))
    else:
        return float(frac)

monsters_df['cr'] = monsters_df['cr'].apply(convert)
monsters_df.head()
```

```python
#Check dtypes and convert if needed
monsters_df.dtypes
```

```python
#Split speed column on ','
monsters_df[['fly', 'swim']] = monsters_df['speed'].str.split(',', expand=True)
monsters_df = monsters_df.drop(columns='speed')

monsters_df.head(10)
```

```python
#Replace NaN values
monsters_df[['fly', 'swim']] = monsters_df[['fly', 'swim']].fillna(0)
monsters_df.head()
```

```python
#Drop extra strings in fly+swim columns
monsters_df['fly'] = monsters_df['fly'].replace(['fly'], 'Yes')
monsters_df['fly'] = monsters_df['fly'].replace(['swim', 0], 'No')
monsters_df['swim'] = monsters_df['swim'].replace(['swim'], 'Yes')
monsters_df['swim'] = monsters_df['swim'].replace(['fly', 0], 'No')
monsters_df.head()
```

```python
#Export final DataFrame
monsters_df.to_csv("cleaned_dnd_monsters.csv", encoding='utf-8', index=False)
```
